```python
import json
import time

with open('data.json') as f:
    data = json.load(f)

datenow = time.strftime("%H:%M-%d.%m.%Y")
filename = "file_" + datenow + ".txt"

with open(filename, "w") as f:
    for name, option, value in [("starea_vremii", "sourceMap", "ftrue"),
                                 ("apa", "preserveConstEnums", "true"),
                                 ("ploaia", "noImplicitAny", "strue")]:
        print(f'"{name}":')
        f.write(f'"{name}":\n')
        for d in data:
            if isinstance(d, dict):
                if option in d and d[option] == value:
                    module_name = d.get("module")
                    print(f'\t{module_name}')
                    f.write(f'\t{module_name}\n')
                elif option not in d:
                    print(f"\t'{option}' key not found in {d}")
                    f.write(f"\t'{option}' key not found in {d}\n")
    print(f"\n-------------------======={datenow}=======--------------------\n")
    f.write(f"\n-------------------======={datenow}=======--------------------\n")
```
